```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
fps = ['power_density_mask', 'wind_speed_mask', 'protected_areas_mask', 'M']
names = ['Power Density', 'Wind Speed', 'Protected Areas', 'All']
data = {name: np.load(os.path.join("masks", fp + '.npy')) for fp, name in zip(fps, names)}
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for ax, name in zip(axes, names):
    ax.imshow(data[name])
    ax.set_title(name)
    ax.axis('off')
plt.tight_layout()
```
